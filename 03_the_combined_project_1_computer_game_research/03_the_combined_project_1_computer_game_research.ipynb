Ниже привожу два варианта решения Задания 8. В обоих вариантах выбираем платформы с суммарными продажами > 500.

1. Вариант. 

Выберем первые 6 платформ из отсортированных данных выше. Они удовлетворяют условию по суммарными продажами > 500.

````python
# С помощью сводной таблицы выберем первые 6 платформ из отсортированных данных выше:
total_top_platforms = games.pivot_table(
    index='platform', values='total_sales', aggfunc='sum').sort_values(by='total_sales', ascending=False
).head(6).index

# Отфильтруем эти платформы в датафрейме games и построим распределение по годам:
games[games['platform'].isin(total_top_platforms)].groupby(
    ['platform', 'year_of_release']
)['total_sales'].sum().unstack('platform').\
plot(grid=True, figsize=(16, 5), title='Распределение платформ с наибольшими суммарными продажами по годам')
````

2. Вариант. 

Используем условие по суммарными продажами > 500 непосредственно в коде.

````python
# Создадим сводную таблицу, в которой отражена информация о названиях платформ и суммарных продажах по каждой из них:
games_top = games.pivot_table(
    index='platform', values='total_sales', aggfunc='sum').sort_values(by='total_sales', ascending=False
)

# Переименуем столбец total_sales таблицы games_pivot:
games_top = games_top.rename(
    columns={
        'total_sales': 'total_sales_platform'
    }
)

# Объединим датафрейм games и сводную талицу games_pivot по столбцу platform:
games_top = games.merge(games_top, how='left', on='platform')

# Выберем платформы с наибольшими суммарными продажами и построим распределение по годам.
# Исходя из информации выше, выберем платформы с суммарными продажами > 500:
games_top[games_top['total_sales_platform'] > 500].groupby(
    ['platform', 'year_of_release']
)['total_sales'].sum().unstack('platform').\
plot(grid=True, figsize=(16, 5), title='Распределение платформ с наибольшими суммарными продажами по годам')
````

В качестве среднего жизненного цикла платформы будем считать 2 года. Данные за 2014-2016 гг. помогут построить прогноз на 2017 год. Данные за предыдущие годы учитывать не будем.

**Задание 11.** Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграмму рассеяния за период c 2014 по 2016 гг. и посчитаем корреляцию между отзывами и продажами. Сформулируем выводы.

### Проверим гипотезы

**Задание 16.** Средние пользовательские рейтинги платформ *Xbox One* и *PC* за период c 2014 по 2016 гг. одинаковые.

```
H_0: Средний пользовательский рейтинг (user_score) платформы *Xbox One* = средний пользовательский рейтинг (user_score) платформы *PC*.
H_a: Средний пользовательский рейтинг (user_score) платформы *Xbox One* ≠ средний пользовательский рейтинг (user_score) платформы *PC*.
alpha = 0.05
```

````python
# Удалим пропуски в столбце user_score

games_user_score = actual_games.dropna(subset=['user_score']).reset_index(drop=True)
games_user_score['user_score'].isna().sum()
````

````python
# Проверим уникальные значения в столбце platform (платформа Xbox One внесена под названием XOne)

games_user_score['platform'].unique()
````

````python
from scipy import stats as st
````

````python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    games_user_score[games_user_score['platform'] == 'XOne']['user_score'], 
    games_user_score[games_user_score['platform'] == 'PC']['user_score'],
    equal_var = False)

# alpha = задайте значение уровня значимости
alpha = .05

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
````

**Задание 17.** Средние пользовательские рейтинги жанров *Action* (англ. «действие», экшен-игры) и *Sports* (англ. «спортивные соревнования») за период c 2014 по 2016 гг. одинаковые.

```
H_0: Средний пользовательский рейтинг (user_score) жанра *Action* = средний пользовательский рейтинг (user_score) жанра *Sports*.
H_a: Средний пользовательский рейтинг (user_score) жанра *Action* ≠ средний пользовательский рейтинг (user_score) жанра *Sports*.
alpha = 0.05
```

````python
# Проверим уникальные значения в столбце genre (все значения совпадают с названиями в гипотезах)

games_user_score['genre'].unique()
````

````python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    games_user_score[games_user_score['genre'] == 'Action']['user_score'], 
    games_user_score[games_user_score['genre'] == 'Sports']['user_score'],
    equal_var = False)

# alpha = задайте значение уровня значимости
alpha = .05

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
````

Для p-значений обеих нулевых гипотез не имеет значение, какой критический уровень статистической значимости для их проверки мы выберем (пороговое значение alpha = 0,01 или 0,05). P-значение для первой нулевой гипотезы = 0,12, а для второй = $1,18*10^{-14}$. Первая нулевая гипотеза в любом случае не будет отвергнута, а вторая - будет.
